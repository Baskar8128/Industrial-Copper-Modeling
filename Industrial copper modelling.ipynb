```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Copper_Set.csv')
data=pd.DataFrame(df)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# heatmap for missing values
sns.heatmap(data.isnull(),cbar = False)
plt.title('Heatmap of missing data', fontsize = 15)
plt.xlabel('Variable', fontsize = 15)
plt.ylabel('material_ref', fontsize = 15)
plt.show()
```

```python
#replacing missing values with most common value
data['material_ref'] = data['material_ref'].fillna(data['material_ref'].mode()[0])
```

```python
#converting inappropriate values to null value
data.loc[data['material_ref'].str.startswith('0000'), 'material_ref']=None
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
sns.boxplot(data)
```

```python
#using isolation forest method to remove outliers
from sklearn.ensemble import IsolationForest
col_name = 'selling_price'
data1 = IsolationForest(n_estimators=100, contamination=0.01)
data1.fit(data[[col_name]])
out_labels = data1.predict(data[[col_name]])
df1 = data[out_labels==1]
```

```python
#histogram for data visualization
df1.hist(bins=10,figsize=(15,10))
plt.show()
```

```python
corr = df1.corr()
corr
```

```python
fig,ax=plt.subplots(figsize=(15,10))
sns.heatmap(df1.corr(),annot=True,ax=ax)
```
